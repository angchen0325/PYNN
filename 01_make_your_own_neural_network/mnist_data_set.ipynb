```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams["font.family"] = "Helvetica"
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
data_file = open("mnist-dataset/mnist_train_100.csv", 'r')
data_list = data_file.readlines()
data_file.close()
```

```python
digit_index = 31 # from 0 to 99
all_values = data_list[digit_index].split(',')
image_array = np.asfarray(all_values[1:]).reshape((28,28))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4.5, 4.5))
im = ax.imshow(
    image_array,
    # extent=[-1, 1, 1, -1],
    # origin="lower",
    aspect="auto",
    cmap="Greys",  # 'rainbow', 'RdBu', 'RdBu_r', 'coolwarm'
)

ax.set_aspect("equal")
# plt.colorbar(im, shrink=1)
# ax.set_xlabel(r"$\hat{k}_x=\sin\theta\cos\phi$")
# ax.set_ylabel(r"$\hat{k}_y=\sin\theta\sin\phi$")
# ax.set_title(f"Pupil spectra with overlay = {ol*1e3:.0f} nm")

fig.tight_layout()
```
